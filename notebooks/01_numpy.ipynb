## Importamos la libreria 
Numpy que es la librería fundamental de Python para el cálculo científico y análisis de datos.

```python
# Importamos la librería NumPy y utilizamos alias como np.
# Se utiliza para trabajar con arreglos, matrices y operaciones matemáticas.
import numpy as np
```

## CARGA DEL ARCHIVO CSV

```python

# np.genfromtxt permite leer datos incluso si hay errores o valores vacíos
# El uso de genfromtxt() proporciona una forma segura y flexible de importar información
# ya que permite trabajar incluso cuando el archivo contiene celdas vacías o registros incompletos.
cantidad, precio_unitario = np.genfromtxt(
    'ventas_techventas.csv',

    # Indica que está separado por comas
    delimiter=',',

    # Salta la fila de encabezados
    skip_header=1,

    # Nos permite solo seleccionar columnas necesarias
    # columna 7 = cantidad
    # columna 8 = precio_unitario
    usecols=(7, 8),

    # unpack=True separa columnas en variables
    # cantidad, precio_unitario
    unpack=True,

    # Evita que el programa se caiga si hay filas dañadas
    invalid_raise=False,

    # Codificación del archivo
    encoding='latin1',

    # Si hay celdas vacías o datos inválidos los convierte en NaN 
    filling_values=np.nan
)

print ("Cantidad: " ,cantidad)
print()
print("Precio uniatrio: ", precio_unitario)
```

## LIMPIEZA DE DATOS

```python
# Convertir negativos a positivos primero
cantidad = np.abs(cantidad)
precio_unitario = np.abs(precio_unitario)

# Una máscara es un arreglo de valores verdaderos que se utiliza 
# para filtrar o seleccionar elementos específicos de una colección de datos. 
mascara_validos = (

    # Para identificar registros con valores nulos 
    # Se aplicó el operador ~ para conservar únicamente los registros que contienen información válida
    ~np.isnan(cantidad) &
    ~np.isnan(precio_unitario) &

    # Validacion de datos negativos 
    (cantidad >= 0) &
    (precio_unitario >= 0)
)

# Aplicar limpieza
cantidad_limpia = cantidad[mascara_validos]
precio_limpio = precio_unitario[mascara_validos]

print ("Cantidad limpia: " ,cantidad_limpia)
print()
print("Precio uniatrio limpio : ", precio_limpio)
```

Ahora nuestros datos estan limpios

# NumPy & Fundamentos Python

### Álgebra Lineal Aplicada
- Crea arrays de cantidades y precios del dataset.
- Calcula el vector de ingresos brutos usando multiplicación matricial con numpy.dot o el operador @.
- Compara con la multiplicación elemento a elemento.

```python
#Calculamos el vector de ingresos brutos usando multiplicación matricial con numpy.dot
ingresos = np.dot(cantidad_limpia, precio_limpio)

print("Ingreso total: ", ingresos)
```

```python
#Calculamos el vector de ingresos por venta con multiplicacion de elemento a elemento 
ingreso_elemento = cantidad_limpia * precio_limpio

print("Ingresos por venta: ", ingreso_elemento)
```

## Comparación de métodos

La multiplicación matricial usando .dot() realiza el producto punto entre ambos vectores, devolviendo un único valor que corresponde al ingreso bruto total de todas las ventas.

Por otro lado, la multiplicación elemento a elemento genera un vector donde cada posición representa el ingreso bruto de una venta individual.

Por tanto:

- Multiplicación matricial → ingreso bruto total del conjunto de ventas. 

- Multiplicación elemento a elemento → ingresos individuales por venta.

## Estadísticas con NumPy
Calcula con NumPy puro sobre el campo precio_unitario:
- Media
- Mediana 
- Desviación estándar 
- Percentiles 25 y 75
Agrega un bloque Markdown explicando qué te dicen estos números sobre el catálogo.

```python
# Media
media = np.mean(precio_limpio)

# Mediana
mediana = np.median(precio_limpio)

# Desviación estándar
desviacion = np.std(precio_limpio)

# Percentiles 25 y 75
percentiles = np.percentile(precio_limpio, [25, 75])

print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacion)
print("Percentil 25:", percentiles[0])
print("Percentil 75:", percentiles[1])
```

## Interpretación de las estadísticas

La media del precio unitario es de 378.25 , lo que indica que en el promedio, los productos del catálogo tienen un costo relativamente alto. 

Sin embargo, la mediana es de 120, mucho menor que la media, lo que sugiere que la mayoría de los productos tienen precios bajos o moderados y que existen algunos productos de precio elevado, que incrementan el promedio general

La desviación estándar de 445.06 es alta, lo que evidencia una gran variabilidad en los precios del catálogo. Esto significa que existen diferencias importantes entre los productos ofrecidos

El percentil 25 (92.5) indica que el 25% de los productos tienen precios iguales o inferiores a este valor, mientras que el percentil 75 (350 ) muestra que el 75% de los productos tienen precios iguales o inferiores a 350. Esto confirma que la mayoría de los productos se concentra en rangos de precio bajos y medios, aunque existen algunos artículos considerablemente más costosos que generan una distribución asimétrica de los precios.
